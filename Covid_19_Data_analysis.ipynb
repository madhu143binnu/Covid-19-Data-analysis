The are some column names which are long, contains special
characters, spaces etc. Rename all such column names accordingly.
Example Country/Region → country New cases → New_cases

```python
from pyspark.sql.types import *

schema = StructType([
    StructField("Country/Region", StringType()),
    StructField("Confirmed", IntegerType()),
    StructField("Deaths", IntegerType()),
    StructField("Recovered", IntegerType()),
    StructField("Active", IntegerType()),
    StructField("New cases", IntegerType()),
    StructField("New deaths", IntegerType()),
    StructField("New recovered", IntegerType()),
    StructField("Deaths / 100 Cases", FloatType()),
    StructField("Recovered / 100 Cases", FloatType()),
    StructField("Deaths / 100 Recovered", FloatType()),
    StructField("Confirmed last week", IntegerType()),
    StructField("1 week change", IntegerType()),
    StructField("1 week % increase", FloatType()),
    StructField("WHO Region", StringType())
])
```

```python
df_01 = df.withColumnRenamed('Country/Region','Country'). \
                             withColumnRenamed('New cases','New_cases').\
                             withColumnRenamed('New deaths','New_death').\
                             withColumnRenamed('New recovered','New_recovered').\
                             withColumnRenamed('Deaths / 100 Cases','Deaths_Cases').\
                             withColumnRenamed('Recovered / 100 Cases','Recovered_Cases').\
                             withColumnRenamed('Deaths / 100 Recovered','Death_Recovered').\
                             withColumnRenamed('Confirmed last week','Confimred_last_week').\
                             withColumnRenamed('1 week change','1_week_change').\
                             withColumnRenamed('1 week % increase ','1_week_%Increase').\
                             withColumnRenamed('WHO Region','WHO_Region')
df_01.show(truncate=False)
df_01.printSchema()
```

```python
country_wise_latest_df = spark.read.csv(path='country_wise_latest.csv',
                             header=False,
                             sep="|",
                             schema=schema,
                             quote="'")
country_wise_latest_df.printSchema()
country_wise_latest_df.show()
```
